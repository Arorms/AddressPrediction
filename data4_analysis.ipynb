# IPv6地址分析 (2001:4开头)

```python
import pandas as pd
import matplotlib.pyplot as plt
from tqdm import tqdm
import ipaddress

# 文件路径
file_path = 'give_data/4_give.txt'
```

## 1. 基本信息统计

```python
def count_lines(filename):
    with open(filename) as f:
        return sum(1 for _ in f)

total_lines = count_lines(file_path)
print(f"总行数: {total_lines:,}")
```

## 2. 前缀分布分析

```python
def analyze_prefixes(filename, sample_size=10000):
    prefix_counts = {}
    
    with open(filename) as f:
        for line in tqdm(f, total=total_lines):
            ip = line.strip()
            try:
                prefix = ':'.join(ip.split(':')[:4])
                prefix_counts[prefix] = prefix_counts.get(prefix, 0) + 1
            except:
                continue
    
    return pd.DataFrame.from_dict(prefix_counts, orient='index', columns=['count'])

prefix_df = analyze_prefixes(file_path)
prefix_df.sort_values('count', ascending=False, inplace=True)
prefix_df.head(20)
```

```python
plt.figure(figsize=(12, 6))
prefix_df.head(20).plot(kind='bar')
plt.title('Top 20 IPv6前缀分布')
plt.xlabel('IPv6前缀')
plt.ylabel('出现次数')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## 4. 地址类型分析

```python
def analyze_address_types(filename):
    types = {'global': 0, 'private': 0, 'other': 0}
    
    with open(filename) as f:
        for line in tqdm(f, total=total_lines):
            ip = line.strip()
            try:
                addr = ipaddress.IPv6Address(ip)
                if addr.is_global:
                    types['global'] += 1
                elif addr.is_private:
                    types['private'] += 1
                else:
                    types['other'] += 1
            except:
                continue
    
    return pd.DataFrame.from_dict(types, orient='index', columns=['count'])

type_df = analyze_address_types(file_path)
type_df.plot(kind='pie', y='count', autopct='%1.1f%%', figsize=(8, 8))
plt.title('IPv6')
plt.show()
```
